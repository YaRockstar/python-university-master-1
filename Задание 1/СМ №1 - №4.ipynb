```python
#  2.
import math


def task2():
    x = float(input(" x ( ): "))
    
    denominator = (math.sin(x) - math.cos(x)) ** 2
    if math.isclose(denominator, 0.0, abs_tol=1e-12):
        print(":    (sin(x) = cos(x))")
        return

    y = (1 + math.sin(2 * x)) / denominator
    print(f' : {y}')


task2()
```

```python
#  5.
def task5():
    x1, y1 = map(int, input("    (x1 y1): ").split())
    x2, y2 = map(int, input("    (x2 y2): ").split())

    cell1 = (x1, y1)
    cell2 = (x2, y2)

    for x, y in (cell1, cell2):
        if not (1 <= x <= 8 and 1 <= y <= 8):
            print(":      [1, 8]")
            return

    if cell1 == cell2:
        print(":    ")
        return

    can_move = max(abs(cell1[0] - cell2[0]), abs(cell1[1] - cell2[1])) == 1
    result = f"      {cell1}   {cell2}"
    result = "" + result if can_move else "" + result
    print(result)


task5()
```

```python
#  1.
def task1():
    summ = 100
    count = 100

    bull_cost = 10
    bulls_max = summ // bull_cost
    
    cow_cost = 5
    cows_max = summ // cow_cost
    
    calve_cost = 0.5
    calves_max = summ // calve_cost
    
    solutions = []
    for bulls in range(bulls_max + 1):
        for cows in range(cows_max + 1):
            calves = count - bulls - cows
            if calves < 0:
                continue

            result_sum = bull_cost * bulls + cow_cost * cows + calve_cost * calves
            if result_sum == summ:
                solutions.append((bulls, cows, calves))

    for bulls, cows, calves in solutions:
        print(f": {bulls}, : {cows},  {calves}")


task1()
```

```python
#  2.
import math


def recursive_sin(x, n):
    if n == 1:
        return math.sin(x)

    return math.sin(recursive_sin(x, n - 1))


def task2():
    n = int(input("   n: "))
    if n < 1:
        print(":  n     ")
        return
    
    x = float(input("   x: "))
    s1 = 0
    s2 = 0
    
    term = x
    for i in range(1, n + 1):
        s1 += recursive_sin(x, i)
        term = math.sin(term)
        s2 += term
    
    print(f'  (): {s1}')
    print(f'  (): {s2}')
    
    
task2()
```

```python
#  3.

#  (2n!) / n!.
def special_fact(n):
    if n <= 1:
        return 2

    result = 1
    for number in range(n + 1, 2 * n + 1):
        result *= number

    return result


def task3():
    epsilon = 0.001
    n = 1
    result = 0

    while True:
        calc = 1 / special_fact(n)
        if calc < epsilon:
            break
        result += calc
        n += 1

    print(f' : {result}')


task3()
```

```python
#  4.
from decimal import Decimal, ROUND_HALF_UP


def task4():
    n = int(input("   N: "))
    if n <= 0:
        print(":  N   ")
        return

    print("    ", n)
    numbers = [Decimal(input(f" {i}  : ")) for i in range(1, n + 1)]
    
    rounded = [int(number.quantize(Decimal("1"), rounding=ROUND_HALF_UP)) for number in numbers]
    print(" :", *rounded)
    print(":", sum(rounded))

    
task4()
```

```python
#  1.
def is_friendly_numbers(first_number, second_number):
    if first_number == second_number:
        return False
    
    first_number_sum = 0
    for i in range(1, first_number // 2 + 1):
        if first_number % i == 0:
            first_number_sum += i
        
    second_number_sum = 0
    for i in range(1, second_number // 2 + 1):
         if second_number % i == 0:
            second_number_sum += i
    
    return first_number_sum == second_number and second_number_sum == first_number
                  

def task1():
    n = int(input("  N: "))
    if n <= 0:
        print(":  N   ")
        return

    friendly_numbers = set()
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            if is_friendly_numbers(i, j):
                pair = tuple(sorted((i, j)))
                friendly_numbers.add(pair)
    
    print(f'  " "   {n} : {len(friendly_numbers)}')
    print(*friendly_numbers)
        

task1()
```

```python
#  2.
import math


def calc_regular_triangle_area(side):
    height = (math.sqrt(3) / 2) * side
    return 0.5 * side * height


def task2():
    side = 6
    result_area = side * calc_regular_triangle_area(side)
    print("  :", result_area)


task2()
```

```python
#  3.
def simm(s, i, j):
    if i >= j:
        return True
    
    return s[i] == s[j] and simm(s, i + 1, j - 1)


def task3():
    original_str = input(" : ")
    clear_str = "".join(char.lower() for char in original_str if char.isalpha() or char.isdigit())
    is_palindrome = simm(clear_str, 0, len(clear_str) - 1)
    print(f" {original_str} - {'' if is_palindrome else ' '}")


task3()
```
